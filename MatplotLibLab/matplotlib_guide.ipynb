2. สร้าง กราฟ plot จะเป็น line graph

```python
import matplotlib.pyplot as plt
import numpy as np

# กำหนดแกน X และ Y ด้วย Numpy array 

x = np.array([2018,2019,2020,2021,2022,2023,2024])
y = np.array([1050,2145,3454,1010,2045,4053,5053])

# สร้าง กราฟ ด้วย matplotlib

plt.plot(x,y)
plt.xticks(x)
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

3. การวาด Bar Graph จะใช้คำสั่งด้านล่าง

```python

plt.bar(x,y)
plt.xticks(x)
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

4. การโหลด data จาก CSV มา plot graph

```python
import matplotlib.pyplot as plt
import pandas as pd

# สร้าง DataFrame จากไฟล์ CSV 
df = pd.read_csv('datasets/gdp.csv')
#df

plt.plot(df['year'],df['gdp']) #สร้างเส้น กราฟ
plt.xticks(df['year']) # กำหนดแกน X ด้วยค่าใน Column Year

plt.xlabel('Year') # สร้าง Label แกน X 
plt.ylabel('TH GDP') # สร้าง Label แกน Y
plt.show() # คำสั่ง Plot Graph
```

```python
# bar chart

plt.bar(
    df['year'],
    df['gdp'],
    color = ['green','red','blue','yellow','pink'],
    edgecolor = 'gray'
)
plt.xticks(df['year']) # กำหนดแกน X ด้วยค่าใน Column Year

plt.xlabel('Year') # สร้าง Label แกน X 
plt.ylabel('TH GDP') # สร้าง Label แกน Y
plt.show() # คำสั่ง Plot Graph
```

```python
# pie chart 

plt.pie(
    df['gdp'],
    labels = df['year'],
    autopct= '%.2f%%',
    colors= ['green', 'red', 'blue', 'yellow','lightgreen']
)
plt.title('Thailand GDP')
plt.show()
```

```python
# Scatter Plot วิธีการ Plot

plt.scatter(df['year'],df['gdp']) #สร้างเส้น กราฟ
plt.xticks(df['year']) # กำหนดแกน X ด้วยค่าใน Column Year
plt.title('Thailand GDP') # กำหนด ชื่อกราฟ 
plt.xlabel('Year') # สร้าง Label แกน X 
plt.ylabel('TH GDP') # สร้าง Label แกน Y
plt.show() # คำสั่ง Plot Graph
```
